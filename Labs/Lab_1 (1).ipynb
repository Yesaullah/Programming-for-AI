```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

choice = input("Enter the function you want to perform: ")

if choice == "addition":
  print("Result: ", num1 - (-num2))
elif choice == "subtraction":
  print("Result: ", num1 - num2)
elif choice == "multiply":
  print("Result: ", num1*num2)
elif choice == "division":
  print("Result: ", num1/num2)
else:
  print("Invalid choice!\nChoose from: 'addition', 'subtraction', 'multiplication', 'division'")
```

# **Task 5:**

image.png

```python
n = int(input("Enter the number of elements: "))
target = int(input("Enter the target integer: "))

nums = []

for i in range(0, n):
    num = int(input("Enter element {}: ".format(i+1)))
    nums.append(num)

print("Initial List: ", nums)

nums = [element for element in nums if element > target]
print("Final List: ", nums)
```

```python
physics = int(input("Enter obtained marks in Physics: "))
chemistry = int(input("Enter obtained marks in Chemistry: "))
maths = int(input("Enter obtained marks in Maths: "))

marks = {
    "Physics": physics,
    "Chemistry": chemistry,
    "Maths": maths
}

sum = 0
highest = 0

for key, value in marks.items():
  sum += value
  if value > highest:
    subject = key
    highest = value

print("Average Marks: ", sum/3)
print(f"Highest marks are scored in {subject}")
```

# **Task 7:**

image.png

```python
word = input("Enter a word: ")
letters_list = list(word)

final_list = []

for i in range(len(letters_list) - 1, -1, -1):
  final_list.append(letters_list[i])

print("Reversed string: ", ''.join(final_list))
```

```python
total_marks = int(input("Enter the total marks for each exam: "))

physics_marks = int(input("Enter the marks obtained in Physics: "))
chemistry_marks = int(input("Enter the marks obtained in Chemistry: "))
maths_marks = int(input("Enter the marks obtained in Maths: "))

marks = {
    "Physics": physics_marks,
    "Chemistry": chemistry_marks,
    "Maths": maths_marks
}

sum = 0

for key, value in marks.items():
  sum += value
  print("Percentage (Physics)", (value/total_marks)*100)

print("The average marks are: ", sum/3)
```
