# **Practice Task 3:**

```python
def find_maximum(*numbers):
    max = 0
    for number in numbers:
        if number > max:
            max = number
    return max

num1 = 10
num2 = 30
num3 = 20

print("Max of the three numbers is: ", find_maximum(num1, num2, num3))
```

# **Task 1:**

```python
import math

def areaTrapezoid(base1, base2, height):
    return 0.5 * (base1 + base2) * height

def areaParallelogram(base, height):
    return base * height

def volumeCylinder(radius, height):
    return math.pi * radius ** 2 * height

def surfaceareacylinder(radius, height):
    return 2 * math.pi * radius * (radius + height)

trapezoid_area = areaTrapezoid(10, 15, 7)
parallelogram_area = areaParallelogram(12, 8)
cylinder_vol = volumeCylinder(5, 10)
cylinder_area = surfaceareacylinder(5, 10)

print("Area of the trapezoid:", trapezoid_area)
print("Area of the parallelogram:", parallelogram_area)
print("Volume of the cylinder:", cylinder_vol)
print("Surface area of the cylinder:", cylinder_area)
```

# **Task 5:**


```python
def Factorial(n):
    if n < 0:
        return "Factorial of negative number does not exist"
    if n == 0:
        return 1
    else:
        return n * Factorial(n - 1)

n = int(input("Enter a number: "))
print(Factorial(n))
```

```python
import random

temperatures = [random.randint(25, 40) for _ in range(30)]
temperatures.sort()
print(temperatures)

sum = 0
for temp in temperatures:
    sum += temp

print(f"The average temperature for the month is: {sum/30}")
print(f"The highest temperature for the month is: {temperatures[-1]}")
print(f"The lowest temperature for the month is: {temperatures[0]}")

# removing the record of the 3rd day
del temperatures[2]
print(temperatures)
```

# **Task 11:**


```python
students={
    'Shadab' :['A','A','B'],
    'Ehtisham': ['A','B','C'],
    'Hasnain': ['B', 'A', 'D']
    }
grade_scale = {
    'A': 4.0,
    'B': 3.0,
    'C': 2.0,
    'D': 1.0,
    'F': 0.0
}
print("Current Students: ")
print(students)

print(("Options: \n1) Adding a Grade (Press '1')\n2) Calculate Average Grade (Press '2')\n3) Add New Student (Press '3')\n4) Remove a Student (Press '4')"))
choice = input("Enter Your Choice: ")
if choice == "1":
    name=input('Enter the Name of the Student: ')
    grade=input("Enter the grade (A,B,C,D,F): ")
    if(name in list(students.keys())):
        List=list(students[name.capitalize()])
        List.append(grade)
        students[name.capitalize()]=List
        print(students)
    else:
        print("Student Not Found")
elif choice == "2":
    name=input('Enter the Name of the Student: ')
    name=name.capitalize()
    if(name in list(students.keys())):
        List=students[name.capitalize()]
        j=0
        for i in List:
            List[j] =grade_scale[i]
            j+=1
        avg=sum(List)/len(List)
        avg =float(round(avg))
        key_list = list(grade_scale.keys())
        val_list = list(grade_scale.values())

        position = val_list.index(avg)
        print(f"The average grade of {name} is {key_list[position]}")
    else:
        print("Student Not Found")
elif choice == "3":
    name=input("Enter the Name of the Student: ")
    students[name.capitalize()]=[]
    print(students)
elif choice == "4":

    name=input("Enter the Name of the Student You Want to Remove: 2")
    name=name.capitalize()
    if(name in list(students.keys())):
        del students[name.capitalize()]
        print("Student ",name," deleted")
    else:
        print("Student Not Found")
```
