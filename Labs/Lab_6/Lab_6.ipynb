```python
orders = pd.read_csv(r'/content/new_orders.csv')
products = pd.read_csv(r'/content/new_products.csv')
```

```python
products.head(5)
```

```python
orders.head(5)
```

```python
products.tail(10)
```

```python
orders.tail(10)
```

```python
# merging the two dataframes
merged_df = pd.merge(orders, products, on='ProductID')
merged_df.head(5)
```

```python
merged_df['Revenue'] = merged_df['Price']*merged_df['Quantity']
print("Total revenue generated: $", merged_df['Revenue'].sum())
```

```python
product_sales = merged_df.groupby('ProductID')['Quantity'].sum().reset_index()
product_sales.head(5)
```

```python
# Merge with products_df to get product names
product_sales = pd.merge(product_sales, products[['ProductID', 'ProductName']], on='ProductID')
product_sales.head(5)
```

```python
best_selling = product_sales.sort_values(by='Quantity', ascending=False).head(5)
low_selling = product_sales.sort_values(by='Quantity', ascending=True).head(5)
```

```python
print("Top 5 best selling ptoducts:")
best_selling
```

```python
print("Top 5 low selling products:")
low_selling
```

```python
best_selling_with_category = pd.merge(best_selling, products[['ProductID', 'Category']], on='ProductID')
best_selling_with_category.head(5)
```

```python
most_common_category = best_selling_with_category['Category'].mode()[0]
print(f"The most common product category among the top 5 best-selling products is: {most_common_category}")
```

```python
category_avg_price = products.groupby('Category')['Price'].mean().reset_index()
```

```python
# making a new dataframe by grouping the the data in profucts by the category
category_avg_price = products.groupby('Category')

category_avg_price = category_avg_price['Price'].mean().reset_index()

category_avg_price
```

```python
# Converting 'Order Date' to datetime
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'])

# making new columns for each part of the date
merged_df['Day'] = merged_df['Order Date'].dt.day
merged_df['Month'] = merged_df['Order Date'].dt.month
merged_df['Year'] = merged_df['Order Date'].dt.year

# finding day, month, and year with max revenues
day_revenue = merged_df.groupby('Day')['Revenue'].sum().idxmax()
month_revenue = merged_df.groupby('Month')['Revenue'].sum().idxmax()
year_revenue = merged_df.groupby('Year')['Revenue'].sum().idxmax()

print(f"Day with highest revenue: {day_revenue}")
print(f"Month with highest revenue: {month_revenue}")
print(f"Year with highest revenue: {year_revenue}")
```

```python
monthly_revenue = merged_df.groupby('Month')['Revenue'].sum().reset_index()

print("Total revenue for each month:")
monthly_revenue
```
