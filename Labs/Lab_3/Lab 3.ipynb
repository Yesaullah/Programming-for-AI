```python
try:
    with open("data.txt", 'r') as file:
        data = file.read()
except FileNotFoundError:
    print("Error: The file 'data.txt' was not found.")
    data = ""
except PermissionError:
    print("Error: You do not have permission to read the file 'data.txt'.")
    data = ""
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
    data = ""

words = data.split()
count = len(words)

print("The total words in the file are:", count)
```

# **Task 2:**


```python
file_data = []

try:
    with open("data2.txt", "r") as file:
        lines = file.readlines()
        for line in lines:
            if "Yesaullah" in line:
                new_line = line.replace("Yesaullah", "Muneeb")
                file_data.append(new_line)
            else:
                file_data.append(line)
except FileNotFoundError:
    print("Error: The file 'data.txt' was not found.")
except PermissionError:
    print("Error: You do not have permission to read the file 'data.txt'.")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")

try:
    with open("data2.txt", 'w') as file:
        file.writelines(file_data)
except PermissionError:
    print("Error: You do not have permission to write to the file 'data.txt'.")
except Exception as e:
    print(f"An error occurred while writing to the file: {e}")
```

# **Task 3:**


```python
n = int(input("Enter the number of elements: "))
key = []
value = []
for i in range(n):
    key.append(input(f"Enter {i+1} element for the first list: "))
    value.append(input(f"Enter {i+1} element for the second list: "))

dict = {}
for i in range(n):
    dict[key[i]] = value[i]

try:
    with open("dict.txt", 'w') as file:
        for key, value in dict.items():
            file.write(f"{key} : {value}\n")
except PermissionError:
    print("Error: Permission to open the file denied.")
except Exception as e:
    print(f"An error occurred: {e}")
```

# **Task 5:**


```python
n = int(input("Enter the number of people: "))
dict = {}
names = []
ages = []
for i in range(n):
    names.append(input("Enter the name of the person: ")) 
    ages.append(int(input("Enter the age of the person: "))) 
    dict[names[i]] = ages[i]

import json

try:
    with open('people.json', 'w') as file:
        json.dump(dict, file, indent=4)
except PermissionError:
    print("Permission denied to write to the file.")
except Exception as e:
    print("An error occurred: ", str(e))

max = 0
c = 0
for i in range(0, len(ages)):
    if ages[i] > max:
        max = ages[i]
        c = i
print(f"{names[c]} has the maximum age: {max}")

for i in range(0, len(ages)):
    if i < len(ages):
        for j in range(i+1, len(ages)):
            if ages[i] == ages[j]:
                print(f"{names[i]} and {names[j]} are of the same age")
```

```python
def questionWriter():
    question = input("Enter a question: ")
    if question.endswith("?"):
        try:
            with open("questions.txt", 'a') as file:
                file.write(question)
        except PermissionError:
            print("You do not have permission to write to this file.")
        except IsADirectoryError:
            print("This is a directory, not a file.")
        except Exception as e:
            print(f"An error occurred: {e}")
    else:
        print("The statement is not a question")

# calling the defined function
questionWriter()
```

# **Task 7:**


```python
try:
    with open('replacement_needed.txt', 'r') as file:
        lines = file.read()
except FileNotFoundError:
    print("File not found")
except IOError:
    print("Error reading file")
except Exception as e:
    print(f"An error occurred: {e}")
else:
    print("No errors.")

print("The content of the file is: ")
print(lines)

mistake = input("Enter the letter you want to replace: ")
correction = input("Enter the replacement letter: ")

lines = lines.replace(mistake, correction)

try:
    with open('replacement_needed.txt', 'w') as file:
        file.write(lines)
except PermissionError:
    print("Permission denied to write to this file.")
except IsADirectoryError:
    print("This is a directory, not a file.")
except Exception as e:
    print(f"Error encountered: {e}")
else:
    print("No errors.")
```
